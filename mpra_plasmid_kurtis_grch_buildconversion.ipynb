```python
def filter_df_for_gene1_query(df, df2):
    t = time.process_time()

    df_copy = df.copy()
    eqtl_list = []
    co_eqtl_list = []
    df_list = []
    df_coeqtl_list = []
    eqtl= list(df2['eqtl'])
    coeqtl = list(df2['co_eqtl'])
    snp = list(df2[ 'snp_id'])
    for eqtl, coeqtl, snp in zip(eqtl, coeqtl, snp):
        df_eqtl = df_copy.query('snp == @snp and gene == @eqtl')
        if not df_eqtl.empty:
            eqtl_list.append(df_eqtl)
            df_sc_eqtl =df2.query('snp_id == @snp and eqtl == @eqtl and co_eqtl==@coeqtl')#and co_eqtl== @coeqtl')
            df_list.append(df_sc_eqtl)
        else:
            df_coeqtl = df_copy.query('snp == @snp and gene == @coeqtl')
            co_eqtl_list.append(df_coeqtl)
            df_sc_coeqtl = df2.query('snp_id == @snp and co_eqtl==@coeqtl and eqtl==@eqtl')
            df_coeqtl_list.append(df_sc_coeqtl)
            if df_coeqtl.empty and df_eqtl.empty:
                raise ValueError('No eqtl or co_eqtl present, {snp} {eqtl}, {coeqtl}')
                print(snp, eqtl, coeqtl)
    elapsed_time = time.process_time() - t        
    eqtl = pd.concat(eqtl_list)
    coeqtl = pd.concat(co_eqtl_list)
    df_eqtl = pd.concat(df_list)
    df_coeqtl = pd.concat(df_coeqtl_list)
    print(elapsed_time, 'seconds')
    return eqtl, coeqtl, df_eqtl, df_coeqtl
```
