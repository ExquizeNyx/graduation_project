```python
def subset_dataframe(df):
    """
    Create subsets of the dataframe based on the chromosome entry and store the subsets into a dictionary
    Based on the key create seperate dataframes into variables
    df = dataframe
    """
    if 'chromosome' not in df.columns:
        raise ValueError('There is no chromosome column in the dataframe')

    #Create list with unique key values
    chromosome_list = list(df['chromosome'].unique())
    df_dict = {}
    for chr in chromosome_list:
        #Create subset of dataframe
        new_df = df[df['chromosome'] == chr]
        #store dataframe with chromosomes as key into a dict
        df_dict[chr] = new_df
    return df_dict
```

```python
def dataframe_to_bed(dataframe, location, name):
    df = dataframe.copy()
    path = location + '/' + name + '.txt'
    df = df[['snp_id', 'snp_chromosome', 'snp_position']].astype({'snp_chromosome': 'str', 'snp_position': 'int64'})
    #df['snp_chromosome'] = 'chr' + df['snp_chromosome']
    end_pos = [i + 1 for i in df['snp_position']]
    df['end'] = end_pos
    df_qtl = df[['snp_chromosome', 'snp_position','end', 'snp_id']].astype({'snp_position': 'int64', 'end': 'int64'} )
    df_qtl.to_csv(path, header=None, sep='\t', index=False)
    return df_qtl

```

```python
def dataframe_to_bed(dataframe, location, name):
    df = dataframe.copy()
    path = location + '/' + name + '.txt'
    df = df[['snp_id', 'snp_chromosome', 'snp_position', 'chr_snp_pos']].astype({'snp_chromosome': 'str', 'snp_position': 'int64'})
    #df['snp_chromosome'] = 'chr' + df['snp_chromosome']
    end_pos = [i + 1 for i in df['snp_position']]
    df['end'] = end_pos
    df_qtl = df[['snp_chromosome', 'snp_position','end','snp_id', 'chr_snp_pos']].astype({'snp_position': 'int64', 'end': 'int64'} )
    df_qtl.to_csv(path, header=None, sep='\t', index=False)
    return df_qtl

```
